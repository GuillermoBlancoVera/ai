# K-Means Clustering on Iris

K-Means is one of the most common clustering algorithms. This notebook clusters the Iris dataset after scaling and evaluates the structure with silhouette score and adjusted Rand index.

## Imports and data

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.metrics import adjusted_rand_score, silhouette_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

iris = load_iris(as_frame=True)
x = iris.data
known_labels = iris.target

x.head()
```

## Fit K-Means

```python
model = make_pipeline(
    StandardScaler(),
    KMeans(n_clusters=3, n_init=10, random_state=42),
)

clusters = model.fit_predict(x)
scaled_x = model.named_steps["standardscaler"].transform(x)

metrics = {
    "silhouette_score": silhouette_score(scaled_x, clusters),
    "adjusted_rand_index": adjusted_rand_score(known_labels, clusters),
}

metrics
```

## Cluster profile

```python
profile = x.assign(cluster=clusters).groupby("cluster").mean()
profile
```

## Practical note

K-Means is fast and useful when clusters are roughly round and similarly sized. It requires choosing the number of clusters up front.
