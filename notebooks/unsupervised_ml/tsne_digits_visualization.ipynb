# t-SNE Visualization on Digits

t-SNE is a nonlinear dimensionality reduction method commonly used to visualize high-dimensional data. This notebook embeds a subset of handwritten digit images into two dimensions.

## Imports and data

```python
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

digits = load_digits()
x = digits.data[:600]
labels = digits.target[:600]
scaled_x = StandardScaler().fit_transform(x)

x.shape
```

## Fit t-SNE

```python
model = TSNE(
    n_components=2,
    perplexity=30,
    init="pca",
    learning_rate="auto",
    random_state=42,
)

embedding = model.fit_transform(scaled_x)
embedding_df = pd.DataFrame(embedding, columns=["tsne_1", "tsne_2"])
embedding_df["digit"] = labels

embedding_df.head()
```

## Embedding spread by digit

```python
embedding_df.groupby("digit")[["tsne_1", "tsne_2"]].mean().round(2)
```

## Practical note

t-SNE is best for visualization and neighborhood exploration, not as a general-purpose preprocessing step. Distances between far-apart groups should be interpreted carefully.
