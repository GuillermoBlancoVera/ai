# Agglomerative Clustering on Iris

Agglomerative clustering builds clusters bottom-up by repeatedly merging nearby groups. This notebook applies hierarchical clustering to Iris and compares linkage strategies.

## Imports and data

```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import load_iris
from sklearn.metrics import adjusted_rand_score, silhouette_score
from sklearn.preprocessing import StandardScaler

iris = load_iris(as_frame=True)
x = iris.data
known_labels = iris.target
scaled_x = StandardScaler().fit_transform(x)

x.head()
```

## Compare linkage methods

```python
results = []

for linkage in ["ward", "complete", "average", "single"]:
    model = AgglomerativeClustering(n_clusters=3, linkage=linkage)
    clusters = model.fit_predict(scaled_x)
    results.append(
        {
            "linkage": linkage,
            "silhouette_score": silhouette_score(scaled_x, clusters),
            "adjusted_rand_index": adjusted_rand_score(known_labels, clusters),
        }
    )

pd.DataFrame(results).sort_values("silhouette_score", ascending=False)
```

## Fit selected model

```python
model = AgglomerativeClustering(n_clusters=3, linkage="ward")
clusters = model.fit_predict(scaled_x)

x.assign(cluster=clusters).groupby("cluster").mean()
```

## Practical note

Agglomerative clustering is useful when hierarchical structure matters. It is more interpretable than many clustering methods, but can be slower on large datasets.
