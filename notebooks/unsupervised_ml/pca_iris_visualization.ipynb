# PCA Visualization on Iris

Principal Component Analysis reduces feature dimensions while preserving as much variance as possible. This notebook projects Iris to two components and reports explained variance.

## Imports and data

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

iris = load_iris(as_frame=True)
x = iris.data

x.head()
```

## Project to two components

```python
model = make_pipeline(StandardScaler(), PCA(n_components=2))
components = model.fit_transform(x)
pca = model.named_steps["pca"]

projection = pd.DataFrame(components, columns=["pc1", "pc2"])
projection["species"] = iris.target_names[iris.target]

summary = {
    "pc1_explained_variance": pca.explained_variance_ratio_[0],
    "pc2_explained_variance": pca.explained_variance_ratio_[1],
    "total_explained_variance": pca.explained_variance_ratio_.sum(),
}

summary
```

## Practical note

PCA is useful for compression, visualization, noise reduction, and feature exploration. Because it is linear, it may miss nonlinear structure.
