# Gaussian Mixture Model on Iris

Gaussian Mixture Models use probabilistic cluster assignments instead of hard distance-only partitions. This notebook fits a GMM and inspects both labels and membership probabilities.

## Imports and data

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.metrics import adjusted_rand_score, silhouette_score
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler

iris = load_iris(as_frame=True)
x = iris.data
known_labels = iris.target
scaled_x = StandardScaler().fit_transform(x)

x.head()
```

## Fit Gaussian Mixture

```python
model = GaussianMixture(n_components=3, covariance_type="full", random_state=42)
clusters = model.fit_predict(scaled_x)
probabilities = model.predict_proba(scaled_x)

metrics = {
    "silhouette_score": silhouette_score(scaled_x, clusters),
    "adjusted_rand_index": adjusted_rand_score(known_labels, clusters),
    "bic": model.bic(scaled_x),
}

metrics
```

## Membership probabilities

```python
pd.DataFrame(probabilities, columns=["component_0", "component_1", "component_2"]).head()
```

## Practical note

Gaussian Mixture Models are useful when clusters overlap and soft assignment is valuable. They assume each component can be modeled with a Gaussian distribution.
