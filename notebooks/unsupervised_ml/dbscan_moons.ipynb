# DBSCAN on Two Moons

DBSCAN finds dense regions and can label isolated points as noise. This notebook uses a synthetic two-moons dataset where density-based clustering is more appropriate than round-cluster methods.

## Imports and data

```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

x, _ = make_moons(n_samples=400, noise=0.08, random_state=42)
scaled_x = StandardScaler().fit_transform(x)

features = pd.DataFrame(scaled_x, columns=["feature_1", "feature_2"])
features.head()
```

## Fit DBSCAN

```python
model = DBSCAN(eps=0.28, min_samples=6)
labels = model.fit_predict(scaled_x)

cluster_count = len({label for label in labels if label != -1})
noise_count = int((labels == -1).sum())
mask = labels != -1

summary = {
    "clusters_found": cluster_count,
    "noise_points": noise_count,
    "silhouette_score_without_noise": silhouette_score(scaled_x[mask], labels[mask]),
}

summary
```

## Cluster counts

```python
pd.Series(labels, name="cluster").value_counts().sort_index()
```

## Practical note

DBSCAN is useful when clusters have irregular shapes and when noise detection matters. It depends heavily on `eps` and the scale of the features.
