# Gradient Boosting Classification

Gradient boosting builds an ensemble sequentially, with each tree correcting errors from earlier trees. This notebook trains a compact boosted model and evaluates it on a binary classification task.

## Imports and data

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer(as_frame=True)
x = cancer.data
y = cancer.target

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)

x.head()
```

## Train and evaluate

```python
model = GradientBoostingClassifier(random_state=42)
model.fit(x_train, y_train)

predictions = model.predict(x_test)
print(f"Accuracy: {accuracy_score(y_test, predictions):.3f}")
print(classification_report(y_test, predictions, target_names=cancer.target_names))
```

## Feature importance

```python
pd.Series(model.feature_importances_, index=x.columns).sort_values(ascending=False).head(10)
```

## Practical note

Gradient boosting is a strong choice for tabular classification when accuracy matters. It has more tuning knobs than a random forest, especially learning rate, tree depth, and number of estimators.
