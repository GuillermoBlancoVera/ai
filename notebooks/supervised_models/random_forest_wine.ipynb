# Random Forest Classification on Wine

Random forests combine many decision trees to reduce overfitting and improve accuracy. This notebook trains a compact forest and inspects feature importance.

## Imports and data

```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split

wine = load_wine(as_frame=True)
x = wine.data
y = wine.target

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)

x.head()
```

## Train and evaluate

```python
model = RandomForestClassifier(n_estimators=200, random_state=42)
model.fit(x_train, y_train)

predictions = model.predict(x_test)
print(f"Accuracy: {accuracy_score(y_test, predictions):.3f}")
print(classification_report(y_test, predictions, target_names=wine.target_names))
```

## Feature importance

```python
pd.Series(model.feature_importances_, index=x.columns).sort_values(ascending=False).head(8)
```

## Practical note

Random forests are reliable tabular-data baselines because they capture nonlinear patterns with little preprocessing. They are usually less interpretable than a single tree.
