# Support Vector Machine Classification

Support Vector Machines are powerful supervised classifiers, especially with scaled numeric features. This notebook trains an RBF-kernel SVM on the breast cancer dataset.

## Imports and data

```python
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

cancer = load_breast_cancer(as_frame=True)
x = cancer.data
y = cancer.target

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)

x.head()
```

## Train and evaluate

```python
model = make_pipeline(StandardScaler(), SVC(kernel="rbf", C=1.0, gamma="scale"))
model.fit(x_train, y_train)

predictions = model.predict(x_test)
print(f"Accuracy: {accuracy_score(y_test, predictions):.3f}")
print(classification_report(y_test, predictions, target_names=cancer.target_names))
```

## Compare kernels

```python
for kernel in ["linear", "rbf"]:
    candidate = make_pipeline(StandardScaler(), SVC(kernel=kernel, gamma="scale"))
    candidate.fit(x_train, y_train)
    print(kernel, f"accuracy={candidate.score(x_test, y_test):.3f}")
```

## Practical note

SVMs can perform very well on medium-sized numeric datasets. They are sensitive to scaling and can become expensive on very large datasets.
