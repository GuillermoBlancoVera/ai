# Linear Regression on Diabetes

Linear regression is a core supervised model for predicting continuous values. This notebook fits a baseline regression model and evaluates prediction error.

## Imports and data

```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

diabetes = load_diabetes(as_frame=True)
x = diabetes.data
y = diabetes.target

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

x.head()
```

## Train and evaluate

```python
model = LinearRegression()
model.fit(x_train, y_train)

predictions = model.predict(x_test)
metrics = {
    "mae": mean_absolute_error(y_test, predictions),
    "rmse": mean_squared_error(y_test, predictions) ** 0.5,
    "r2": r2_score(y_test, predictions),
}

metrics
```

## Inspect coefficients

```python
pd.Series(model.coef_, index=x.columns).sort_values(key=abs, ascending=False)
```

## Practical note

Linear regression is a strong first model when the target is numeric and interpretability matters. Its coefficients are easy to inspect, but it can underfit nonlinear relationships.
